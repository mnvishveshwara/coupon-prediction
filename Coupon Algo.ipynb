```python
import pandas as pd

path="C:\\Users\\Vishva\\OneDrive\\Desktop\\python\\Data\\Coupon Predictor\\Coupons.csv"
data=pd.read_csv(path)

print(data)
print(data.info())


import sklearn
from sklearn.preprocessing import LabelEncoder
le_Gender=LabelEncoder()
le_Item1=LabelEncoder()
le_Item2=LabelEncoder()
le_Item3=LabelEncoder()
le_Coupon=LabelEncoder()



data['Gender_n']=le_Gender.fit_transform(data['Gender'])                    # encoding values
data['Item1_n']=le_Item1.fit_transform(data['Item1'])
data['Item2_n']=le_Item2.fit_transform(data['Item2'])
data['Item3_n']=le_Item3.fit_transform(data['Item3'])
data['Coupon_n']=le_Coupon.fit_transform(data['Coupon'])


print(data)


inputs = data.drop(['Coupon','Coupon_n','Gender','Item1','Item2','Item3','Gender_n'],'columns')
output = data.drop(['Gender','Age','Item1','Item2','Item3','Coupon','Item1_n','Item2_n','Item3_n','Gender_n'],'columns')
print("input:",inputs)
print("output:",output)




```

```python
import sklearn
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(inputs,output,train_size=0.8)

print(x_train)
print(x_test)
print(y_train)
print(y_test)

from sklearn.svm import SVC
model = SVC()                         #initialize the model
model.fit(x_train,y_train)            #training
y_pred = model.predict(x_test)
print(y_pred)
print(y_test)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)     #actual values , predicted values
print(cm)


res=model.predict([[40,0,1,4]])                             #predict the output
print("result:",res)

acc=model.score(inputs,output)*100
print("acc:",acc)
```
